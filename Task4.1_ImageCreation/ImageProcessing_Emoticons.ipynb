```python
#Task 4.1
import numpy as np
import cv2
```

```python
#TEXT
text = np.zeros((500,500,3))
font = cv2.FONT_HERSHEY_SIMPLEX
org = (40, 250)
fontScale = 1
color = (0, 255, 255)
thickness = 2
image = cv2.putText(text, 'Expression- Sad to Happy!', org, font, fontScale, color, thickness, cv2.LINE_AA)
```

```python
#showImage

cv2.imshow("my photo",text)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#SAD
sad = np.zeros((500,500,3))
# main_circle_face
center_coordinates = (250,250)
radius = 150
color = (0,255,255)
thickness = -2
photo = cv2.circle(sad,center_coordinates,radius,color,thickness)

#left_eye 
center_coordinates = (200, 200)  
axesLength = (15, 7)
angle = 90
startAngle = 0
endAngle = 360
color = (0, 0, 0)
thickness = -5
image = cv2.ellipse(sad, center_coordinates, axesLength,
           angle, startAngle, endAngle, color, thickness)

#Right_eye
center_coordinates = (300, 200)  
axesLength = (15, 7)
angle = 90
startAngle = 0
endAngle = 360
color = (0, 0, 0)
thickness = -5
image = cv2.ellipse(sad, center_coordinates, axesLength,
           angle, startAngle, endAngle, color, thickness)

#sad
center_coordinates = (250, 300)  
axesLength = (80, 60)
angle = 0
startAngle = 180
endAngle = 360
   
# Red color in BGR
color = (0, 0,0)
thickness = 5
   
# Using cv2.ellipse() method
# Draw a ellipse with red line borders of thickness of 5 px
image = cv2.ellipse(sad, center_coordinates, axesLength,
           angle, startAngle, endAngle, color, thickness)
```

```python
#showImage

cv2.imshow("my photo",sad)
cv2.waitKey()
cv2.destroyAllWindows()
```
